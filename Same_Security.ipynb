### Cleaning the text

#### removing the stopwords and unneccesory text

```python

lemmatizer = WordNetLemmatizer()

def rm_stopwords(text):

   j = re.sub('[^a-zA-Z0-9]',' ',text)
   # print(j)
   j = j.lower()
   # print('snd : ',j)
   j = j.split()
   # print('3rd',j)
   j = [lemmatizer.lemmatize(word) for word in j if word not in stopwords.words('english')]
   return ' '.join(j)



def updated_df(dataframe,column_name):

  dataframe[column_name] = dataframe[column_name].apply(rm_stopwords)
  return dataframe


```

####  Word2Vect

we need the combined text

```python
train_new['combined_desc'] = train_new['x'] + ' ' + train_new['y']
train_final = train_new.drop(['x','y'],axis=1)
train_final.head()
```

```python
!pip install gensim
```

Here we need the sentences

```python
train_final['tockens'] = train_final['combined_desc'].apply(lambda x: x.split())
```

```python
train_final['tockens']
```

Initialize and train the model

```python
import gensim
from gensim.models.word2vec import Word2Vec

w2v = Word2Vec(sentences=train_final['tockens'],
               vector_size=100,
               min_count=5,
               sg=1,        # skipgram for semantic
               window = 5)
```

```python
w2v.wv
```

create vectors

```python
import numpy as np

def sentence_vectors(sentence,model,vector_size):

  vectors = [model.wv[word] for word in sentence if word in model.wv]

  if(len(vectors) == 0):
    return np.zeros(vector_size)

  return np.mean(vectors,axis=0)
```

Implimetnting vectors & define features

```python
X_w2v = np.vstack(
    train_final['tockens'].apply(lambda x: sentence_vectors(x,w2v,100))
)
y = train_final['same_security'].astype(int)
```

```python
X_w2v
```

split the dataset

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split( X_w2v,y,test_size=0.2,random_state=42,stratify=y)
```

Initialize the model

```python
from sklearn.neighbors import KNeighborsClassifier

knc = KNeighborsClassifier(n_neighbors=4)
```

Train

```python
knc.fit(X_train,y_train)
```

Predict

```python
y_pred_w2v = knc.predict(X_test)
```

Classification report


```python
from sklearn.metrics import classification_report

print(classification_report(y_pred_w2v,y_test))
w2v_class_report = classification_report(y_pred_w2v,y_test)
```
